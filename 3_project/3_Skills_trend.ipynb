```python
# Importing Libraries
import ast
import pandas as pd
import seaborn as sns 
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA_US = df[(df['job_title'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()
```

```python
df_DA_US['job_posted_month_no'] = df_DA_US['job_posted_date'].dt.month
```

```python
df_DA_US_explode = df_DA_US.explode('job_skills')
```

```python
# Create the pivot table
df_DA_US_pivot = df_DA_US_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)

# Add the 'Total' row
df_DA_US_pivot.loc['Total'] = df_DA_US_pivot.sum()

# Sort the columns by the 'Total' row in descending order
df_DA_US_pivot = df_DA_US_pivot[df_DA_US_pivot.loc['Total'].sort_values(ascending=False).index]

# Drop the 'Total' row
df_DA_US_pivot = df_DA_US_pivot.drop('Total')

# Display the sorted pivot table
df_DA_US_pivot
```

```python
df_DA_US_pivot.plot(kind='line')
# Temporary chart. Not useful - updated below. 
```

```python
DA_total = df_DA_US.groupby('job_posted_month_no').size()
```

```python
df_DA_US_percent = df_DA_US_pivot.div(DA_total/100, axis=0)



df_DA_US_percent
```

```python
df_plot = df_DA_US_percent.iloc[:, :5]

sns.lineplot(data=df_plot, dashes=False, palette='tab10') 
sns.set_theme(style='ticks')
sns.despine()

plt.title('Trending top Skills for Data Analyst in the US monthly (Jan-Dec)')
plt.ylabel('Likelihood in Job Posting')
plt.xlabel('2023')
plt.legend().remove()

from matplotlib.ticker import PercentFormatter
ax = plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))


for i in range(5):
    plt.text(12.3, df_plot.iloc[-1, i],df_plot.columns[i])
```
